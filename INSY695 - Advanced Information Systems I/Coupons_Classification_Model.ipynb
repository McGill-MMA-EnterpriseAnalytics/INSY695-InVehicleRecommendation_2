### Classification Model Building

```python
## Importing machine learning libraries

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
## Splitting data into train & test datasets
X = df_og.iloc[:,1:]
y = df_og['Y']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=5)
```

```python
## Creating GBT function

def gbt_thresh(X_train,X_test,y_train,y_test):   
    
    gbt = GradientBoostingClassifier() 
    model = gbt.fit(X_train, y_train)
    
    y_test_pred = model.predict(X_test)
    
    #accuracy_score(y_test, y_test_pred)

    # Calculate the accuracy score
    y_test_pred = model.predict(X_test)
    r1 = metrics.accuracy_score(y_test, y_test_pred)

    # Print the confusion matrix
    metrics.confusion_matrix(y_test, y_test_pred)
    # Calculate the Precision/Recall
    r2 = metrics.precision_score(y_test, y_test_pred)
    r3 = metrics.recall_score(y_test, y_test_pred)
    # Calculate the F1 score
    r4 = metrics.f1_score(y_test, y_test_pred)
    
    return r1, r2, r3, r4
```

```python
## Creating Logistic Regression function

def logistic_reg(X_train,X_test,y_train,y_test):
    
    lr2 = LogisticRegression()
    model = lr2.fit(X_train,y_train)
    
    y_test_pred = model.predict(X_test)
    #accuracy_score(y_test, y_test_pred)

    # Calculate the accuracy score
    y_test_pred = model.predict(X_test)
    r5 = metrics.accuracy_score(y_test, y_test_pred)

    # Print the confusion matrix
    metrics.confusion_matrix(y_test, y_test_pred)
    # Calculate the Precision/Recall
    r6 = metrics.precision_score(y_test, y_test_pred)
    r7 = metrics.recall_score(y_test, y_test_pred)
    # Calculate the F1 score
    r8 = metrics.f1_score(y_test, y_test_pred)
    
    return r5, r6, r7, r8
```

```python
## KNN

def knn_algo(X_train,X_test,y_train,y_test):
    
    standardizer = StandardScaler()
    X_train_std = standardizer.fit_transform(X_train)
    X_test_std = standardizer.fit_transform(X_test)
    
    a = []
    for i in range (1,20):
        knn3 = KNeighborsClassifier(n_neighbors=i)
        model3 = knn3.fit(X_train_std,y_train)
        y_test_pred = model3.predict(X_test_std)
        a.append(metrics.accuracy_score(y_test, y_test_pred))
        
    r9 = max(a)
    r10 = metrics.f1_score(y_test, y_test_pred)
        
    return r9, r10
```

```python
## Random Forest

def rf_algo(X_train,X_test,y_train,y_test):
    
    randomforest = RandomForestClassifier()
    model = randomforest.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    
    r11 = metrics.accuracy_score(y_test, y_test_pred)
    r12 = metrics.f1_score(y_test, y_test_pred)
    
    return r11, r12
```

```python
## CART

def cart(X_train,X_test,y_train,y_test):

    a = []
    
    for i in range(2,21):
        decisiontree = DecisionTreeClassifier(max_depth=i)
        model = decisiontree.fit(X_train, y_train)
        y_test_pred = model.predict(X_test)
        a.append(metrics.accuracy_score(y_test, y_test_pred))
    
    r13 = max(a)
    r14 = metrics.f1_score(y_test, y_test_pred)
    
    return r13, r14
```

```python
## Running GBT, Logistic Regression, KNN, Random Forest, CART before feature selection

r1, r2, r3, r4 = gbt_thresh(X_train,X_test,y_train,y_test)
r5, r6, r7, r8 = logistic_reg(X_train,X_test,y_train,y_test)
r9, r10 = knn_algo(X_train,X_test,y_train,y_test)
r11, r12 = rf_algo(X_train,X_test,y_train,y_test)
r13, r14 = cart(X_train,X_test,y_train,y_test)

print("Results before feature selection:")
print("GBT Results:")
print("Accuracy:", r1)
print("F1:", r4)

print("\nLogistic Regression Results:")
print("Accuracy:", r5)
print("F1:", r8)

print("\nKNN Results:")
print("Accuracy:", r9)
print("F1:", r10)

print("\nRandom Forest Results:")
print("Accuracy:", r11)
print("F1:", r12)

print("\nCART:")
print("Accuracy:", r13)
print("F1:", r14)
```

```python
############# Feature Selection ############# 

############# Recursive Feature Elimination for feature selection

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
rfe = RFE(lr, n_features_to_select=1)

model = rfe.fit(X, y)
model.ranking_
rfe_pred = pd.DataFrame(list(zip(X.columns,model.ranking_)), columns = ['predictor','ranking'])
rfe_pred.sort_values(by="ranking", inplace=True)
rfe_pred.reset_index(drop=True, inplace=True)
rfe_pred.head(10)
```

```python
############# Running loop for finding the optimal threshold for RFE feature selection

temp_a = rfe_pred['predictor']
temp_b = []
thresh = []
accuracy_score = []
f1_score = []

for i in range(len(rfe_pred)):
    temp_b.append(temp_a[i])
    
    X = df_og[temp_b]
    r11, r12 = rf_algo(X_train,X_test,y_train,y_test)
    thresh.append(i)
    accuracy_score.append(r11)
    f1_score.append(r12)
    
df_rf_RFE_thresh = pd.DataFrame(list(zip(thresh, accuracy_score, f1_score)), columns=['Threshold', 'Accuracy Score', 'F1 Score'])
df_rf_RFE_thresh

print("Highest Accuracy Score values:")
print(df_rf_RFE_thresh[df_rf_RFE_thresh["Accuracy Score"]==max(df_rf_RFE_thresh["Accuracy Score"])].sort_values(by=['Threshold']))

print("\nHighest F1 Score values:")
print(df_rf_RFE_thresh[df_rf_RFE_thresh["F1 Score"]==max(df_rf_RFE_thresh["F1 Score"])].sort_values(by=['Threshold']))
```

```python
############# LASSO for feature selection

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

from sklearn.linear_model import Lasso

temp_a = []
temp_b = [x / 1000 for x in range(10, 50, 1)]
temp_c = []

for i in temp_b:
    model = Lasso(alpha=i)
    model.fit(X_std,y)
    temp_a = pd.DataFrame(list(zip(X.columns,model.coef_)), columns = ['predictor','coefficient'])
    temp_a = list(temp_a[temp_a['coefficient']!=0]['predictor'])
    temp_c.append(temp_a)

temp_d = [list(x) for x in set(tuple(x) for x in temp_c)]

for i in temp_d:
    temp_e = df_og[i]
    y_ls = df_og['Y']
    X_ls = df_og.iloc[:,1:]
    
    X_train_ls,X_test_ls,y_train_ls,y_test_ls = train_test_split(X_ls,y_ls,test_size=0.3, random_state=5)
    
    r11, r12 = rf_algo(X_train_ls,X_test_ls,y_train_ls,y_test_ls)
    accuracy_score.append(r11)
    f1_score.append(r12)
    
df_rf_Lasso_thresh = pd.DataFrame(list(zip(accuracy_score, f1_score)), columns=['Accuracy Score', 'F1 Score'])
df_rf_Lasso_thresh

print("Highest Accuracy Score values:")
print(df_rf_Lasso_thresh[df_rf_Lasso_thresh["Accuracy Score"]==max(df_rf_Lasso_thresh["Accuracy Score"])])

print("\nHighest F1 Score values:")
print(df_rf_Lasso_thresh[df_rf_Lasso_thresh["F1 Score"]==max(df_rf_Lasso_thresh["F1 Score"])])
```
